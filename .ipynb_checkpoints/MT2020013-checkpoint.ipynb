## Importing modules

```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import statistics as stat
import numpy as np
import seaborn as sns
from collections import Counter
```

## Loading the Dataset 

```python
with open("./facebook.txt") as file:
    data = file.read()
```

## Creating the edge list.

```python
edges = []
for pair in data.split("\n"):
    vertices = pair.split(" ")
    edge = (int(vertices[0]),int(vertices[1]))
    edges.append(edge)
edges
```

## Creating the graph

```python
graph = nx.Graph(edges)
graph
```

## Visualizing the graph

```python
nx.draw(graph, node_size = 2)
```

## Calculating some important properties of this graph

- Size of the graph
- Number of nodes 
- Average degree 
- Average shortest path

Size of the graph

```python
graph.number_of_edges()
```

Number of nodes

```python
graph.number_of_nodes()
```

Average shortest path

```python
nx.average_shortest_path_length(graph)
```

Average degree

```python
stat.mean([graph.degree(n) for n in graph.nodes()])
```

Finding out top 10 most important nodes of the graph and visualize them

- Calculating the degree centrality 
- Calculating the load centrality 
- Calculating the betweenness centrality


```python
nx.spring_layout(graph)

def func(items):
    sorted_items = sorted(items, key = lambda x:x[1], reverse = True)
    sortedCounts = {k:v for k,v in sorted_items}
    impNodes = list(sortedCounts)[:10]
    return impNodes

deg = func(nx.degree_centrality(graph).items())
load = func(nx.load_centrality(graph).items())
between = func(nx.betweenness_centrality(graph).items())
```

```python
print("Top ten nodes by degree centrality :\n",deg)
print("Top ten nodes by load centrality :\n",load)
print("Top ten nodes by betweenness centrality :\n",between)
```

```python
nx.draw(graph, node_color = "yellow", node_size = 2, alpha = 1)
nx.draw(graph, nodelist = deg, node_color = "red", node_size = 30, alpha = 1)
nx.draw(graph, nodelist = load, node_color = "blue", node_size = 30, alpha = 1)
nx.draw(graph, nodelist = between, node_color = "green", node_size = 30, alpha = 1)
```

Finding if it a Scale-free graph.

Let us observe degree distribution curve

```python
degrees = [graph.degree(x) for x in graph.nodes()]
degrees, degrees_frequency = zip(*Counter(degrees).items())
```

```python
plt.figure(figsize=(15,8))
sns.set(style = "darkgrid")
sns.scatterplot(x=degrees, y=degrees_frequency)
plt.title("degree distribution")
plt.xlabel("degree")
plt.ylabel("count")
```

As we can observe the degree distribution follows Power Law and hence we can say this is a scale free network.
